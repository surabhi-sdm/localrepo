```python
from IPython.display import HTML

html_content = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>AI Fitness Coach</title>
    <style>
        body {
            margin: 0;
            font-family: "Poppins", sans-serif;
            background: #f4f7fa;
            color: #333;
        }

        header {
            background: linear-gradient(120deg, #005bea, #00c6fb);
            padding: 40px 20px;
            text-align: center;
            color: white;
        }

        header h1 {
            margin: 0;
            font-size: 36px;
            font-weight: 700;
        }

        header p {
            margin-top: 10px;
            font-size: 18px;
        }

        .container {
            max-width: 900px;
            margin: 40px auto;
            background: white;
            padding: 25px;
            border-radius: 12px;
            box-shadow: 0 6px 20px rgba(0,0,0,0.08);
        }

        h2 {
            border-left: 5px solid #007bff;
            padding-left: 10px;
            font-size: 24px;
            margin-bottom: 20px;
        }

        label {
            display: block;
            font-weight: 600;
            margin: 8px 0 5px;
        }

        input, select, textarea {
            width: 100%;
            padding: 12px;
            border-radius: 8px;
            border: 1px solid #ccc;
            margin-bottom: 15px;
            font-size: 15px;
            outline: none;
            transition: 0.2s;
        }

        input:focus, select:focus, textarea:focus {
            border-color: #007bff;
        }

        .row {
            display: flex;
            gap: 15px;
        }

        .row div {
            flex: 1;
        }

        button {
            width: 100%;
            padding: 15px;
            background: #005bea;
            color: white;
            border: none;
            border-radius: 10px;
            font-size: 18px;
            cursor: pointer;
            margin-top: 20px;
            transition: 0.3s;
        }

        button:hover {
            background: #0042a6;
        }

        .result-box {
            margin-top: 40px;
            padding: 20px;
            border-radius: 12px;
            background: #e8f0ff;
            display: none;
        }

        .result-box h3 {
            font-size: 22px;
        }

        ul {
            margin-top: 10px;
        }

        @media(max-width: 700px) {
            .row { flex-direction: column; }
        }
    </style>
</head>
<body>

<header>
    <h1>AI Fitness Coach</h1>
    <p>Personalized Fitness Plans Tailored to Your Body & Goals</p>
</header>

<div class="container">
    <h2>Your Fitness Profile</h2>

    <label>Name *</label>
    <input id="name" type="text" placeholder="Enter your name">

    <div class="row">
        <div>
            <label>Age</label>
            <input id="age" type="number" placeholder="25">
        </div>
        <div>
            <label>Height (cm)</label>
            <input id="height" type="number" placeholder="170">
        </div>
        <div>
            <label>Weight (kg)</label>
            <input id="weight" type="number" placeholder="70">
        </div>
    </div>

    <div class="row">
        <div>
            <label>Gender</label>
            <select id="gender">
                <option>Male</option><option>Female</option><option>Other</option>
            </select>
        </div>

        <div>
            <label>Fitness Goal</label>
            <select id="goal">
                <option>Overall Fitness</option>
                <option>Weight Loss</option>
                <option>Muscle Gain</option>
            </select>
        </div>

        <div>
            <label>Fitness Level</label>
            <select id="level">
                <option>Beginner</option>
                <option>Intermediate</option>
                <option>Advanced</option>
            </select>
        </div>
    </div>

    <div class="row">
        <div>
            <label>Workout Location</label>
            <select id="location">
                <option>Gym</option>
                <option>Home</option>
                <option>Outdoor</option>
            </select>
        </div>

        <div>
            <label>Dietary Preference</label>
            <select id="diet">
                <option>Non-Vegetarian</option>
                <option>Vegetarian</option>
                <option>Vegan</option>
            </select>
        </div>

        <div>
            <label>Hours per Week</label>
            <input id="hours" type="number" placeholder="5">
        </div>
    </div>

    <label>Medical History (optional)</label>
    <textarea id="medical" placeholder="Any allergies or conditions?"></textarea>

    <label>Stress Level</label>
    <select id="stress">
        <option>Low</option><option selected>Medium</option><option>High</option>
    </select>

    <label>Injuries or Limitations (optional)</label>
    <textarea id="injuries" placeholder="Any physical limitations?"></textarea>

    <button onclick="generatePlan()">Generate My Fitness Plan</button>

    <div class="result-box" id="resultBox">
        <h3>Your Personalized Fitness Plan</h3>
        <p><strong>Name:</strong> <span id="rname"></span></p>
        <p><strong>BMI:</strong> <span id="rbmi"></span></p>
        <p><strong>Goal:</strong> <span id="rgoal"></span></p>

        <h4>Recommended Weekly Plan</h4>
        <ul id="planList"></ul>
    </div>
</div>

<script>
function generatePlan() {
    let name = document.getElementById("name").value;
    let h = parseFloat(document.getElementById("height").value);
    let w = parseFloat(document.getElementById("weight").value);

    let bmi = (w / ((h/100)**2)).toFixed(1);

    document.getElementById("rname").innerText = name;
    document.getElementById("rbmi").innerText = bmi;
    document.getElementById("rgoal").innerText = document.getElementById("goal").value;

    const planList = document.getElementById("planList");
    planList.innerHTML = "";

    const goal = document.getElementById("goal").value;

    let workouts = [];

    if (goal === "Weight Loss") {
        workouts = ["45 min Cardio", "HIIT Training", "Strength + Cardio Mix"];
    } else if (goal === "Muscle Gain") {
        workouts = ["Chest + Triceps", "Back + Biceps", "Leg Day", "Shoulders + Core"];
    } else {
        workouts = ["Full Body Strength", "Cardio Day", "Mobility + Core Stability"];
    }

    workouts.forEach(w => {
        let li = document.createElement("li");
        li.textContent = w;
        planList.appendChild(li);
    });

    document.getElementById("resultBox").style.display = "block";
}
</script>

</body>
</html>"""

HTML(html_content)
```
